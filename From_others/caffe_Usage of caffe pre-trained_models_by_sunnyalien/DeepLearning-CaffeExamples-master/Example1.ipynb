```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from random import randint
%matplotlib inline
```

```python
import sys
sys.path.insert(0, '/home/rna2/caffe/python')
import caffe
```

```python
VIDEO_PATH = '/home/rna2/Seminario/Videos/bird.avi'
MODEL_FILE = '/home/rna2/Seminario/Modelo/cifar10_full.prototxt'
PRETRAINED = '/home/rna2/Seminario/Modelo/cifar10_full_iter_70000.caffemodel'

cap = cv2.VideoCapture(VIDEO_PATH)

caffe.set_mode_cpu()
net = caffe.Classifier(MODEL_FILE, PRETRAINED,
                       channel_swap=None,
                       raw_scale=255)

clases = ["avion", "coche", "pajaro", "gato", "alce", "perro", "rana", "caballo", "barco", "camion"]
```

```python
   
for i in range(10):   
       
    frameNumber = randint(0, 200 )
    cap.set(cv2.cv.CV_CAP_PROP_POS_FRAMES, frameNumber)
    ret, frame = cap.read()       
    
    frame = np.float32(frame)
    frame = frame/255
    b,g,r = cv2.split(frame)
    img = cv2.merge([r,g,b])
        
    prediction = net.predict([img])  # predict takes any number of images, and formats them for the Caffe net automatically
    
    fig = plt.figure()
    plt.subplot(122)   
    plt.imshow(img)
    plt.subplot(121)
    plt.plot(prediction[0])
    plt.show()
    
    
    print 'Clasificado como: '
    print clases[prediction[0].argmax()]
        
       
    
```
